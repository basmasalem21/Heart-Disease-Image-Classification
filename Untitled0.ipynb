```python
import os
import torch
import torch.nn as nn
from torchvision import datasets, models, transforms
from torch.utils.data import DataLoader
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
num_epochs = 10
batch_size = 32
learning_rate = 1e-4
num_classes = 2  # عدد الفئات (مثلاً: "سليم" و"مريض")
```

```python
data_transforms = {
    "train": transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.RandomHorizontalFlip(),
        transforms.RandomRotation(10),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
    ]),
    "test": transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
    ]),
}
```

```python
# ========== تحميل البيانات ==========
train_dataset = datasets.ImageFolder(
    root="/content/drive/MyDrive/heart disease/train/train",
    transform=data_transforms["train"]
)
test_dataset = datasets.ImageFolder(
    root="/content/drive/MyDrive/heart disease/test/test",
    transform=data_transforms["test"]
)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=True)
```

```python
import torch
from torchvision import models
import torch.nn as nn

# تحديد عدد الفئات
num_classes = 10  # عدّل حسب عدد الفئات في بياناتك

# تحميل نموذج MobileNetV3 Small المدرب مسبقًا
model = models.mobilenet_v3_small(pretrained=True)

# استخراج عدد الميزات من الطبقة الأخيرة وتعديلها لتناسب عدد الفئات
num_features = model.classifier[0].in_features  # الطبقة الصحيحة لاستخراج in_features
model.classifier = nn.Sequential(
    nn.Linear(num_features, num_classes)
)

# نقل النموذج إلى الجهاز (GPU أو CPU)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)

#print(model)  # عرض تفاصيل النموذج للتأكد من التعديلات
```

```python
# ========== إعداد التدريب ==========
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

```python
from tqdm import tqdm
# تعريف المتغيرات لتخزين القيم الحقيقية والتنبؤات
all_preds = []
all_labels = []

# ======= تدريب =======
for epoch in range(num_epochs):
    model.train()
    running_loss, correct_train, total_train = 0.0, 0, 0
    print(f"Epoch {epoch + 1}/{num_epochs} - Training...")
    for inputs, labels in tqdm(train_loader, desc="Training Progress"):
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, preds = torch.max(outputs, 1)
        correct_train += (preds == labels).sum().item()
        total_train += labels.size(0)

    train_accuracy = correct_train / total_train
    train_loss = running_loss / len(train_loader)

    # ======= اختبار =======
    model.eval()
    running_loss, correct_test, total_test = 0.0, 0, 0
    print("Testing...")
    with torch.no_grad():
        for inputs, labels in tqdm(test_loader, desc="Testing Progress"):
            inputs, labels = inputs.to(device), labels.to(device)

            outputs = model(inputs)
            loss = criterion(outputs, labels)

            running_loss += loss.item()
            _, preds = torch.max(outputs, 1)
            correct_test += (preds == labels).sum().item()
            total_test += labels.size(0)

            # حفظ التنبؤات والقيم الحقيقية
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    test_accuracy = correct_test / total_test
    test_loss = running_loss / len(test_loader)

    # ======= طباعة النتائج =======
    print(f"Epoch {epoch + 1}/{num_epochs}")
    print(f"Train Loss: {train_loss:.4f}, Train Accuracy: {train_accuracy:.4f}")
    print(f"Test Loss: {test_loss:.4f}, Test Accuracy: {test_accuracy:.4f}")
    print("-" * 50)

```

```python

from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# ======= بعد اكتمال جميع الحقب: إنشاء مصفوفة الارتباك =======
cm = confusion_matrix(all_labels, all_preds)
print("Confusion Matrix:")
print(cm)

# ======= عرض مصفوفة الارتباك =======
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=[f"Class {i}" for i in range(cm.shape[0])], yticklabels=[f"Class {i}" for i in range(cm.shape[0])])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

```

```python
import matplotlib.pyplot as plt

# عدد الصور التي نريد عرضها
num_images_to_show = 5
images_shown = 0

# تأكد من أن النموذج في وضع التقييم
model.eval()

# عرض الصور والتنبؤات
with torch.no_grad():
    for inputs, labels in test_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = model(inputs)
        _, preds = torch.max(outputs, 1)

        # تحويل الصور، التنبؤات، والملصقات إلى وحدة المعالجة المركزية
        inputs_cpu = inputs.cpu()
        labels_cpu = labels.cpu()
        preds_cpu = preds.cpu()

        for i in range(min(num_images_to_show - images_shown, inputs_cpu.size(0))):
            img = inputs_cpu[i]

            # إذا كانت الصورة RGB (3 قنوات)
            if img.shape[0] == 3:
                img = img.permute(1, 2, 0)  # إعادة ترتيب الأبعاد لتصبح (224, 224, 3)

            # عرض الصورة
            plt.imshow(img.numpy(), cmap="gray" if img.ndim == 2 else None)
            plt.title(f"True: {labels_cpu[i].item()}, Pred: {preds_cpu[i].item()}")
            plt.axis("off")
            plt.show()

            images_shown += 1
            if images_shown >= num_images_to_show:
                break

        if images_shown >= num_images_to_show:
            break
```
